## Remarks
- Sorts are done **left to right by convention**, the sort **does not depend on the direction** the sequence is parsed
- **In-Place** means *no extra memory is used* except for temporary variables for swapping
- **Out-of-Place** means *extra memory is used* to store intermediate states
- **Iterative in-place sorting algorithms** do not need to use the `return` statement, but for the sake of testing, it is recommended to include it.
- Only the **Big O Notation** is tested for Time Complexity.

- Swapping always uses a **temporary variable**

> The idiomatic expression for swapping variable values in python is
> 
> ```python
> a, b = b, a
> ```
>
> But really, swapping *always* require an **extra variable** to store the values of one of its variables **temporarily**
> 
> ```python
> temp = a
> a = b
> b = temp
> ```
>
> So actually, the idiomatic expression of `a, b = b, a` is doing this with a temporary variable **under the hood**
>
> In some languages, especially **low-level** ones likes C and C++, swapping will require **explicitly using a temporary variable**

# Bubble Sort

> **In-Place Sorting Algorithm**

A sorting algorithm that repeatedly moves the most **extreme element** to the **boundary of the unsorted portion**

### Method

- Compare each element with its adjacent element and **swap** them if they are in the wrong order.
- Repeat this process for each element until the list is **sorted**.
- Optimize the algorithm by **skipping unnecessary comparisons** and stopping early if the list is already sorted.

### Time Complexity

- Worst-case: **O(n^2)**
- Average-case: **θ(n^2)**
- Best-case: **Ω(n)**


````python
## Un-optimised
def bubblesort_1(nums):
    min_range = len(nums) - 1 

    # Checking all the numbers
    for i in range(min_range):
        # Doing the swaps in the unsorted portion
        for j in range(min_range - i):
            # Swap if curr > next
            if nums[j] > nums[j+1]:
                nums[j], nums[j+1] = nums[j+1], nums[j]
                
    return nums # for easy testing
````

````python
def bubblesort_2(nums):
    min_range = len(nums) - 1
    
    for unsorted in range(min_range, 0, -1):
        for i in range(unsorted):
            if nums[i] > nums[i+1]:
                nums[i], nums[i+1] = nums[i+1], nums[i]
    
    return nums # for easy testing
````

````python
## Optimised
def bubblesort_3(nums):
    min_range = len(nums) - 1
    
    for unsorted in range(min_range, 0, -1):
        # Check for swaps
        swapped = False
        
        for i in range(unsorted):
            if nums[i] > nums[i+1]:
                nums[i], nums[i+1] = nums[i+1], nums[i]
                swapped = True
                
        # Stop if already sorted
        if not swapped:
            break
            
    return nums # for easy testing

## Test
from sorting_utils.sorting_tests import test_sorts
sorts = {
    "bubblesort_1": bubblesort_1,
    "bubblesort_2": bubblesort_2,
    "bubblesort_3": bubblesort_3,
}; test_sorts(sorts)
````

# Insertion Sort

> **In-Place Sorting Algorithm**

A sorting algorithm that moves each element towards the **boundary** of the sequence where they are fitted in the **correct position** within the **continuously growing boundary portion**

### Method
- Let the left side be the **space where sorted elements are accumulated**
- Check each element in the list.
- Compare the **current element** with the **previous element**.
- If the **current element** is smaller than the **previous element**, continuously **shift** the **previous element** backwards until the **correct position** is found within the **sorted portion**

*[Optimization]*: Instead of **swapping** elements, we can **shift** them.

### Time Complexity

- Worst-case: **O(n^2)**
- Average-case: **θ(n^2)**
- Best-case: **Ω(n)**


````python
## Un-optimised
def insertion_sort_1(nums):
    # Check all the numbers (checking from the 2nd one is negligibly faster)
    for i, _ in enumerate(nums):
        
        # Continuously swap if not at first index AND curr < prev
        while (i > 0) and (nums[i] < nums[i-1]):
            nums[i], nums[i-1] = nums[i-1], nums[i]
            i -= 1
    
    return nums # for easy testing

## Optimised
def insertion_sort_2(nums):
    
    for i, curr_num in enumerate(nums):
        while (i > 0) and (curr_num < nums[i-1]):
            # Shift the previous number to the right instead of swapping this time
            nums[i] = nums[i-1]
            i -= 1
            
        # Now we can insert curr_num into correct place
        nums[i] = curr_num
    
    return nums # for easy testing

## Test
from sorting_utils.sorting_tests import test_sorts
sorts = {
    "insertion_sort_1": insertion_sort_1,
    "insertion_sort_2": insertion_sort_2,
}; test_sorts(sorts)
````

# Selection Sort

> **In-Place Sorting Algorithm**

A sorting algorithm that efficiently divides the list into a **sorted left portion** and an **unsorted right portion**. It repeatedly finds the **minimum element** from the unsorted portion and **places it in the sorted portion**.

### Method

- Iterate through each element of the list.
- Find the **minimum element** within the unsorted portion.
- **Swap** the minimum element with the first element in the unsorted portion, thereby expanding the sorted portion while reducing the unsorted portion

### Time Complexity

- Worst-case: **O(n^2)**
- Average-case: **θ(n^2)**
- Best-case: **Ω(n^2)**

````python
def selection_sort(nums):
    # Iterate through the unsorted portion of the list
    for swap_at, _ in enumerate(nums):
        min_at = swap_at

        # Find the index of the minimum element within the unsorted portion
        for i in range(swap_at, len(nums)):
            if nums[i] < nums[min_at]:
                min_at = i

        # If the minimum element is found at a different index, swap them
        if min_at != swap_at:
            nums[min_at], nums[swap_at] = nums[swap_at], nums[min_at]

    return nums


## Test
from sorting_utils.sorting_tests import test_sorts
sorts = {
    "selection_sort": selection_sort,
}; test_sorts(sorts)
````

# Merge Sort

> **Out-of-Place Sorting Algorithm**

Merge sort is a **divide and conquer** algorithm that efficiently sorts large lists by **recursively partitioning** it and **merging** the sorted partitions through **pair-wise comparison**

### Method

- **Recursive Partitioning**
    - Divide the list into **two halves**——recursively——using the **midpoint**.
      - *Ensure the midpoint is calculated correctly to avoid an infinite loop.*
      - i.e. do NOT use `(len(nums) - 1) // 2`, use `len(nums) // 2` instead
    
   
- **Merging**
    - Compare the numbers **pairwise** in the **two sublists** repeatedly, putting the smaller one into a **new list**
    - When **no more comparisons can be done**, the **new list** will take in the **numbers** in the **non-empty sublist** in their **original position**


### Time Complexity

- Worst-case: **O(n log(n))**
- Average-case: **θ(n log(n))**
- Best-case: **Ω(n log(n))**

````python
# Method 1: Better performance for larger lists by using 2 pointers when sorting
def merge_sort_standard(nums):
    
    # Checks for condition where it cannot partition anymore first
    # Also reduces nesting, because the other way is to check len(nums) > 1, which would indent the entire code after this
    # This is called a "guard clause"
    if len(nums) <= 1:
         return nums
        
    # Find mid-point
    mid_at = len(nums) // 2
    
    # Recursive Partition
    left = merge_sort_standard(nums[:mid_at])
    right = merge_sort_standard(nums[mid_at:])
    
    # Sorting
    merged = []
    left_index = 0 
    right_index = 0
    
    # Sorting with 2 pointers
    while (
        left_index < len(left)
        and right_index < len(right)
    ):
        
        # If first number in the left list is smaller than the first number in the right list
        # Add the first number in the left list
        # Update the left index variable
        if left[left_index] < right[right_index]:
            merged.append(left[left_index])
            left_index += 1

        # Otherwise, add the first number in the right list
        # Update the right index variable
        else:
            merged.append(right[right_index])
            right_index += 1
    
    # Put the residual numbers
    # This is an optimisation, we just dump both lists without checking, only one will have elements anyway
    merged.extend(left[left_index:])
    merged.extend(right[right_index:])

    return merged
    

# Method 2: Easier to write, but worse performance on larger lists, i.e thousands or more elements

# This is just another function defined to help the original function
# Thus known as a helper function
def merge(left, right):
    result = []
    
    # Sorting with pop
    while (
        len(left) > 0
        and len(right) > 0
    ):
        if left[0] < right[0]:
            left_num = left.pop(0)
            result.append(left_num)
        else:
            right_num = right.pop(0)
            result.append(right_num)

    # Mandatory Cleaning Operation 
    result.extend(left)
    result.extend(right)
    
    return result

def merge_sort_short(nums):

    # Guard Clause —— refer to explanation in the above code for more details
    if len(nums) <= 1:
        return nums 

    # Find midpoint position
    mid_at = len(nums) // 2
    
    # Recursive Partitioning
    left_partition = merge_sort_short(nums[:mid_at])
    right_partition = merge_sort_short(nums[mid_at:])

    return merge(left_partition, right_partition)


## Test
from sorting_utils.sorting_tests import test_sorts
sorts = {
    "merge_sort_standard": merge_sort_standard,
    "merge_sort_short": merge_sort_short,
}; test_sorts(sorts)
````

# Quick Sort

> **In-Place Sorting Algorithm**

Quick Sort is a **divide and conquer** algorithm that involves selecting a **element** and then doing the same for the 2 subsequences around the **element**

Thus, every iteration sorts the **selected element** into its **correct position**

- This element is thus called the **pivot**
- The repeated division of the sequence around the pivot is called **recursive partitioning** 


## Method

- **Sorting**
    - Find the pivot:
        - First element (**easiest to code, worst performance on large datasets**)
        - Median element (**fast, complicated to code**)
        - Random element (**fast, easy to code**) <--- *I went with this, but its not necessary, just start with whatever is easiest first*
        
    - Compare each number with the pivot and place it on the correct side.
    
    
- **Recursive Partitioning**
    - Splitting the list by the pivot
    
---

### 3-Way Quicksort
> **Out-of-Place Implementation**
- Define three empty lists: `less`, `equal`, `more`, 
- Find the pivot.
- Compare each number with the pivot and sort it into the correct list.
- Perform recursive partitioning.

### Basic Quicksort (Not-in-Syllabus)
> **In-Place Implementation**
- Base case: `start < end`
- Get the pivot and two pointers, `i` and `j`:
    - `i` finds elements greater than the pivot and becomes the next low.
    - `j` finds elements smaller than the pivot and becomes the next high.
- While the pivot is not sorted, update `i` and `j` accordingly.
- Perform recursive partitioning.

### Time Complexity
- Worst-case: **O(n^2)**
- Average-case: **θ(n log(n))**
- Best-case: **Ω(n log(n))**


````python
import random

# Method 1: 3-way Quicksort
def quick_sort_3_way(nums, low=0, high=None):

    # To make "high" a parameter with a default value of len(nums) - 1
    # It cannot be used directly because lists are mutable and default paramters are evaluated once —— when they are first called
    # So the list will not reset after the first time, and build-up elements that are not supposed to be there
    # So this is a work around, have high=None then checking if its None, and updating the parameter
    # Or just not use a parameter at all, and have low and high both defined from within the function
    # It is a design choice
    if high is None:
        high = len(nums) - 1

    # Guard Clause
    if len(nums) <= 1:
        return nums # for easy testing, otherwise, just "return" will do
    
    # Get pivot and 3 empty lists
    pivot = nums[random.randint(low, high)]
    
    less = []
    equal = []
    more = []
    
    # Sort every number in list accordingly
    for num in nums:
        if num < pivot:
            less.append(num)
            
        elif num > pivot:
            more.append(num)
            
        else:
            equal.append(num)
    
    # Recursive Partition
    result = []

    less_sorted = quick_sort_3_way(less)
    right_sorted = quick_sort_3_way(more)
    
    result.extend(less_sorted)
    result.extend(equal)
    result.extend(right_sorted)
    
    return result
        
# Method 2: Basic Quicksort
def quick_sort_basic(nums, low=0, high=None):
    if high is None:
        high = len(nums) - 1

    # Guard Clause
    if low >= high:
        return nums # For easy testing

    # Get pivot
    pivot = nums[random.randint(low, high)]
    
    # Get 2 pointers: i finds greater than pivot, j finds smaller than pivot
    i, j = low, high
    
    # Pivot not sorted
    while i <= j:
        
        # Continuously update both pointers until find right number to swap
        while nums[i] < pivot:
            i += 1
            
        while nums[j] > pivot:
            j -= 1
            
        # If pivot still not sorted
        if i <= j:
            nums[i], nums[j] = nums[j], nums[i]
            i += 1
            j -= 1
            
    # Recursive Partition
    quick_sort_basic(nums, low, j)
    quick_sort_basic(nums, i, high)

    # For easy testing
    return nums
    
## Test
from sorting_utils.sorting_tests import test_sorts
sorts = {
    "quick_sort_3_way": quick_sort_3_way, 
    "quick_sort_basic": quick_sort_basic,
}; test_sorts(sorts)
````
